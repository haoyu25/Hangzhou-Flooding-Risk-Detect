```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from pysheds.grid import Grid
import pysheds
import numpy as np
```

```python
# Merge
import os
import rasterio
from rasterio.merge import merge
import glob

input_folder = "data/raw/dem"
output_path = "data/dem.tif"

dem_files = glob.glob(os.path.join(input_folder, "*.tif"))

src_files_to_mosaic = [rasterio.open(fp) for fp in dem_files]

mosaic, out_trans = merge(src_files_to_mosaic)

out_meta = src_files_to_mosaic[0].meta.copy()

out_meta.update({
    "driver": "GTiff",
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "transform": out_trans,
    "nodata": -9999,
    "dtype": rasterio.float32
})

with rasterio.open(output_path, "w", **out_meta) as dest:
    dest.write(mosaic.astype("float32"))

print(f"finish：{output_path}")
```

```python
# Read raw DEM
demfile = "data/inundation/dem.tif"

grid = Grid.from_raster(demfile)
dem = grid.read_raster(demfile)

# Detect pits
pits = grid.detect_pits(dem)
```

```python
# Fill pits
pit_filled_dem = grid.fill_pits(dem)
pits = grid.detect_pits(pit_filled_dem)
assert not pits.any()
```

```python
# Detect depressions
depressions = grid.detect_depressions(pit_filled_dem)
```

```python
# Fill depressions
flooded_dem = grid.fill_depressions(pit_filled_dem)
depressions = grid.detect_depressions(flooded_dem)
assert not depressions.any()
```

```python
#Resolving flats
inflated_dem = grid.resolve_flats(flooded_dem)
flats = grid.detect_flats(inflated_dem)
# assert not flats.any()
```

```python
# Compute flow direction based on corrected DEM
fdir = grid.flowdir(inflated_dem)

# Compute flow accumulation based on computed flow direction
acc = grid.accumulation(fdir)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
fig.patch.set_alpha(0)
im = ax.imshow(acc, zorder=2,
               cmap='cubehelix',
               norm=colors.LogNorm(1, acc.max()),
               interpolation='bilinear')
plt.colorbar(im, ax=ax, label='Upstream Cells')
plt.title('Flow Accumulation', size=14)
plt.tight_layout()
```

```python
# Compute height above nearest drainage
hand = grid.compute_hand(fdir, dem, acc > 200)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
fig.patch.set_alpha(0)
im = ax.imshow(hand, zorder=2,
               cmap='cubehelix',
               norm=colors.LogNorm(1, acc.max()),
               interpolation='bilinear')
plt.colorbar(im, ax=ax, label='Upstream Cells')
plt.title('Hand', size=14)
plt.tight_layout()
```

```python
import geopandas as gpd
import numpy as np
import rasterio
from rasterio.mask import mask
from matplotlib import colors
import matplotlib.pyplot as plt

boundary = gpd.read_file('data/HZ_Center.shp')

demfile = "data/dem.tif"
with rasterio.open(demfile) as src:
    dem_data = src.read(1)
    transform = src.transform

    geoms = boundary.geometry.values  
    hand_clipped, _ = mask(src, geoms, crop=True)

    hand_clipped = hand_clipped[0]

    meta = src.meta
    meta.update({
        'height': hand_clipped.shape[0],
        'width': hand_clipped.shape[1],
        'transform': transform, 
        'nodata': -9999.0          
    })

    output_file = 'data/hand_clipped.tif'
    with rasterio.open(output_file, 'w', **meta) as dest:
        dest.write(hand_clipped, 1) 
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
fig.patch.set_alpha(0)

im = ax.imshow(hand_clipped, zorder=2, cmap='cubehelix',
               norm=colors.LogNorm(1, hand_clipped.max()),
               interpolation='bilinear')

plt.colorbar(im, ax=ax, label='Height Above Nearest Drainage')
plt.title('Hand (HZ Center)', size=14)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import rasterio
from rasterio import Affine
import os
import matplotlib.pyplot as plt

hand_path = 'data/hand_clipped.tif'
with rasterio.open(hand_path) as src:
    hand = src.read(1)
    profile = src.profile
    nodata = src.nodata
    transform = src.transform

if nodata is not None:
    hand = np.where(hand == nodata, np.nan, hand)

thresholds = [10, 15, 20, 25]# 10 = 5年一遇，15 = 20年一遇，20 = 50年一遇，25 = 100年一遇
fig, axes = plt.subplots(1, len(thresholds), figsize=(20, 5))

for i, t in enumerate(thresholds):
    inundation_extent = np.where(hand < t, t - hand, np.nan)
    img = axes[i].imshow(inundation_extent, cmap='Blues')
    axes[i].set_title(f"HAND threshold: {t}m")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
output_dir = 'data/inundation'
os.makedirs(output_dir, exist_ok=True)

for t in thresholds:
    inundation_extent = np.where(hand < t, t - hand, np.nan).astype('float32')

    meta = profile.copy()
    meta.update({
        'dtype': 'float32',
        'nodata': np.nan,
        'count': 1
    })

    output_path = os.path.join(output_dir, f'inundation_threshold_{t}m.tif')
    
    with rasterio.open(output_path, 'w', **meta) as dst:
        dst.write(inundation_extent, 1)

print("save!:", output_dir)
```

```python
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import rasterio.warp
from pysheds.grid import Grid

shp = gpd.read_file('data/HZ_Center.shp')

with rasterio.open('data/dem.tif') as src:
    if src.crs != shp.crs:
        transform, width, height = rasterio.warp.calculate_default_transform(
            src.crs, shp.crs, src.width, src.height, *src.bounds)
        kwargs = src.meta.copy()
        kwargs.update({'crs': shp.crs, 'transform': transform, 'width': width, 'height': height})

        with rasterio.MemoryFile() as memfile:
            with memfile.open(**kwargs) as temp_dst:
                for i in range(1, src.count + 1):
                    rasterio.warp.reproject(
                        source=rasterio.band(src, i),
                        destination=rasterio.band(temp_dst, i),
                        src_transform=src.transform,
                        src_crs=src.crs,
                        dst_transform=transform,
                        dst_crs=shp.crs,
                        resampling=rasterio.warp.Resampling.nearest)

                out_image, out_transform = mask(temp_dst, shp.geometry, crop=True)
                out_meta = temp_dst.meta.copy()
    else:
        out_image, out_transform = mask(src, shp.geometry, crop=True)
        out_meta = src.meta.copy()

out_meta.update({
    "driver": "GTiff",
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform
})

with rasterio.open('data/inundation/dem_clipped.tif', "w", **out_meta) as dest:
    dest.write(out_image)
```

```python
#Estimating inundation extent
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import numpy as np
import matplotlib.pyplot as plt
from pysheds.grid import Grid

grid = Grid.from_raster('data/inundation/dem_clipped.tif', data_name='dem')
dem = grid.read_raster('data/inundation/dem_clipped.tif', data_name='dem')

threshold = 10
inundation_extent = np.where(hand < threshold, threshold - hand, np.nan)

fig, ax = plt.subplots(figsize=(8,6))
fig.patch.set_alpha(0)
dem_view = grid.view(dem, nodata=np.nan)
plt.imshow(dem_view, extent=grid.extent, cmap='Greys', zorder=1)
plt.imshow(inundation_extent, extent=grid.extent,
           cmap='Blues', vmin=-5, vmax=10, zorder=2)
plt.grid(zorder=0)
plt.title('Inundation depths (Threshold = 10)', size=14)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
```
