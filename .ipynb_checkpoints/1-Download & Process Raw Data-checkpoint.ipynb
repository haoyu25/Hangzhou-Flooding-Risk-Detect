```python
import os
import urllib.request
import rasterio
from rasterio.warp import reproject, Resampling

outfolder = 'data/sentinel-bands-2023'
if not os.path.exists(outfolder): os.makedirs(outfolder)

bands_to_download = ["B02", "B03", "B04", "B08", "B11"] # Blue, Green, Red, NIR, SWIR1

for item in items_tiles:
    band_arrays = []

    for band in bands_to_download:
        href = item.assets[band].href
        
        print(f"Downloading {band} from {href}")
        
        tmp_band_file = os.path.join(outfolder, f"{item.id}_{band}.tif")
        urllib.request.urlretrieve(href, tmp_band_file)

        with rasterio.open(tmp_band_file) as src:
            band_data = src.read(1) 
            band_transform = src.transform 
            band_crs = src.crs 
            
            if band == "B02":
                meta = src.meta  # 

            #  Band 1120m 10m
            if band == "B11":
                #  B02 (10m) 
                reference_shape = band_arrays[0].shape if band_arrays else band_data.shape
                dst_band = np.empty(reference_shape, dtype=band_data.dtype)

                #  10m 
                reproject(
                    source=band_data,
                    destination=dst_band,
                    src_transform=band_transform,
                    src_crs=band_crs,
                    dst_transform=meta['transform'],  #  B02 
                    dst_crs=band_crs,
                    resampling=Resampling.bilinear  # 
                )

                band_arrays.append(dst_band)  #  Band 11 
            else:
                band_arrays.append(band_data)  # 

        os.remove(tmp_band_file)  # 

    # 
    stacked = np.stack(band_arrays, axis=0)  # shape: (5, H, W)

    meta.update(count=5)  # 

    # 
    outfilename = os.path.join(outfolder, f"{item.id}_B2B3B4B8B11.tif")
    with rasterio.open(outfilename, 'w', **meta) as dst:
        dst.write(stacked)  # 

    print(f"Saved 5-band image to {outfilename}")
```

```python
import os, os.path
import numpy as np
import rasterio as rio
import rasterio
from rasterio.plot import show
from rasterio.mask import mask
from rasterio.windows import get_data_window
from rasterio.windows import Window
from rasterio.merge import merge
# optional - turn off warnings
import warnings
from matplotlib import pyplot as plt
warnings.filterwarnings('ignore')
```

```python
import os
import rasterio
from shapely.geometry import box
from itertools import combinations

input_folder = 'data/sentinel-bands-2023'

tif_files = [os.path.join(input_folder, f) for f in os.listdir(input_folder) if f.endswith('.tif')]
print(f"Found {len(tif_files)} tif files.\n")

bounds_list = []
crs_set = set()
res_set = set()

print("Tile Boundaries and Metadata Check:\n")
for tif in tif_files:
    with rasterio.open(tif) as src:
        bounds = src.bounds
        crs = src.crs
        res = src.res
        print(f"{os.path.basename(tif)}")
        print(f"  Bounds: {bounds}")
        print(f"  CRS: {crs}")
        print(f"  Resolution: {res}\n")

        bounds_list.append((os.path.basename(tif), box(*bounds)))
        crs_set.add(crs.to_string())
        res_set.add(res)

print("\nCRS Consistency Check:")
if len(crs_set) > 1:
    print(" Inconsistent CRS found!")
    print(crs_set)
else:
    print(" All CRS are consistent.")

print("\nResolution Consistency Check:")
if len(res_set) > 1:
    print(" Inconsistent resolutions found!")
    print(res_set)
else:
    print(" All resolutions are consistent.")

print("\nTile Overlap Check:")
overlaps = []
for (name1, geom1), (name2, geom2) in combinations(bounds_list, 2):
    if geom1.intersects(geom2):
        inter_area = geom1.intersection(geom2).area
        if inter_area > 0:
            overlaps.append((name1, name2, inter_area))

if overlaps:
    print(f" Found {len(overlaps)} overlapping tile pairs:")
    for a, b, area in overlaps:
        print(f"  - {a} <--> {b}, Overlap Area: {area:.2f}")
else:
    print(" No overlaps detected.")
```

```python
import os
import rasterio
from rasterio.merge import merge
from rasterio.warp import calculate_default_transform, reproject, Resampling
from tqdm import tqdm
import tempfile

input_folder = 'data/sentinel-bands-2023'
output_file = 'data/naip-mosaiced2023.tif'
target_crs = 'EPSG:4326' 

tif_files = [os.path.join(input_folder, f) for f in os.listdir(input_folder) if f.endswith('.tif')]

reprojected_rasters = []

for tif_path in tqdm(tif_files, desc="Reprojecting tiles"):
    with rasterio.open(tif_path) as src:
        if src.crs.to_string() != target_crs:
            transform, width, height = calculate_default_transform(
                src.crs, target_crs, src.width, src.height, *src.bounds)
            kwargs = src.meta.copy()
            kwargs.update({
                'crs': target_crs,
                'transform': transform,
                'width': width,
                'height': height,
                'driver': 'GTiff'  # Ensure the driver is set for MemoryFile
            })

            with tempfile.NamedTemporaryFile(delete=False) as tmpfile:
                tmpfile_path = tmpfile.name
                with rasterio.open(tmpfile_path, 'w', **kwargs) as dst:
                    for i in range(1, src.count + 1):
                        reproject(
                            source=rasterio.band(src, i),
                            destination=rasterio.band(dst, i),
                            src_transform=src.transform,
                            src_crs=src.crs,
                            dst_transform=transform,
                            dst_crs=target_crs,
                            resampling=Resampling.nearest
                        )
                reprojected_rasters.append(tmpfile_path)
        else:
            reprojected_rasters.append(tif_path)

src_files_to_mosaic = [rasterio.open(fp) for fp in reprojected_rasters]
mosaic, out_trans = merge(src_files_to_mosaic)

out_meta = src_files_to_mosaic[0].meta.copy()
out_meta.update({
    "driver": "GTiff",
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "count": mosaic.shape[0],
    "transform": out_trans,
    "crs": target_crs,
    "compress": "lzw",
    "BIGTIFF": "YES"
})

with rasterio.open(output_file, "w", **out_meta) as dest:
    dest.write(mosaic)

print(f"\n Mosaic complete! Output saved to: {output_file}")
```

```python
import fiona
import rasterio
from rasterio.mask import mask
from fiona.transform import transform_geom

input_value_raster = 'data/naip-mosaiced2023.tif'
shpfile = 'data/HZ_Center.shp'
out_raster = 'data/naip2023.tif'

with rasterio.open(input_value_raster) as src:
    raster_crs = src.crs.to_string() 

    # reprojection
    with fiona.open(shpfile, "r") as shapefile:
        shapefile_crs = shapefile.crs['init'] if 'init' in shapefile.crs else shapefile.crs_wkt
        shapes = [
            transform_geom(shapefile_crs, raster_crs, feature["geometry"])
            for feature in shapefile
        ]

        out_image, out_transform = mask(src, shapes, crop=True)
        out_meta = src.meta.copy()

out_meta.update({
    "driver": "GTiff",
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform,
    "compress": 'lzw'
})

with rasterio.open(out_raster, "w", **out_meta) as dest:
    dest.write(out_image)
```

```python
import os, os.path
import numpy as np
import rasterio as rio
import rasterio
from rasterio.plot import show
from rasterio.mask import mask
from rasterio.windows import get_data_window
from rasterio.windows import Window
from rasterio.merge import merge
# optional - turn off warnings
import warnings
from matplotlib import pyplot as plt
warnings.filterwarnings('ignore')
```

```python
import os
import rasterio
from rasterio.merge import merge
from rasterio.warp import calculate_default_transform, reproject, Resampling
from tqdm import tqdm
import tempfile

input_folder = 'data/sentinel-bands-2018'
output_file = 'data/naip-mosaiced2018.tif'
target_crs = 'EPSG:4326' 

tif_files = [os.path.join(input_folder, f) for f in os.listdir(input_folder) if f.endswith('.tif')]

reprojected_rasters = []

for tif_path in tqdm(tif_files, desc="Reprojecting tiles"):
    with rasterio.open(tif_path) as src:
        if src.crs.to_string() != target_crs:
            transform, width, height = calculate_default_transform(
                src.crs, target_crs, src.width, src.height, *src.bounds)
            kwargs = src.meta.copy()
            kwargs.update({
                'crs': target_crs,
                'transform': transform,
                'width': width,
                'height': height,
                'driver': 'GTiff'  # Ensure the driver is set for MemoryFile
            })

            with tempfile.NamedTemporaryFile(delete=False) as tmpfile:
                tmpfile_path = tmpfile.name
                with rasterio.open(tmpfile_path, 'w', **kwargs) as dst:
                    for i in range(1, src.count + 1):
                        reproject(
                            source=rasterio.band(src, i),
                            destination=rasterio.band(dst, i),
                            src_transform=src.transform,
                            src_crs=src.crs,
                            dst_transform=transform,
                            dst_crs=target_crs,
                            resampling=Resampling.nearest
                        )
                reprojected_rasters.append(tmpfile_path)
        else:
            reprojected_rasters.append(tif_path)

src_files_to_mosaic = [rasterio.open(fp) for fp in reprojected_rasters]
mosaic, out_trans = merge(src_files_to_mosaic)

out_meta = src_files_to_mosaic[0].meta.copy()
out_meta.update({
    "driver": "GTiff",
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "count": mosaic.shape[0],
    "transform": out_trans,
    "crs": target_crs,
    "compress": "lzw",
    "BIGTIFF": "YES"
})

with rasterio.open(output_file, "w", **out_meta) as dest:
    dest.write(mosaic)

print(f"\n Mosaic complete! Output saved to: {output_file}")
```
